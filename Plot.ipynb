```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./out/output.csv")
df = df.drop(columns="Entry").multiply(1000)
```

```python
df_melt = pd.melt(df, var_name = "LayerMaterial", value_name='Energy loss (keV)')
grid = sns.FacetGrid(df_melt, col = "LayerMaterial", height=5, col_wrap=3)
grid.map(sns.histplot, "Energy loss (keV)")
```
